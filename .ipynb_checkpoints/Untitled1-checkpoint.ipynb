```python
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_url = 'https://raw.githubusercontent.com/'\
'PacktWorkshops/The-Data-Science-Workshop/'\
'master/Chapter03/bank-full.csv'
bank = pd.read_csv(file_url, sep=";")
```

```python
bank.head()
```

```python
filter_mask = bank['y'] == 'yes'
bank_sub_1 = bank[filter_mask].groupby(["age"])['y'].agg(agegrp='count').reset_index()
bank_sub_1.head()
```

```python
alt.Chart(bank_sub_1).mark_line().encode(x='age', y='agegrp')
```

```python
plt.plot(bank_sub_1['age'], bank_sub_1['agegrp'])
```

```python
age_tot = bank.groupby(['age'])['y'].agg(age_tot='count').reset_index()
age_tot.head()
```

```python
age_categ = bank.groupby(['age', 'y'])['y'].agg(age_categ = 'count').reset_index()
age_categ.head()
```

```python
age_merged = pd.merge(age_categ, age_tot, on='age')
age_merged['age_prop'] = age_merged.age_categ / age_merged.age_tot
```

```python
alt.Chart(age_merged).mark_line().encode(x='age', y='age_prop').facet(column='y')
```

```python
age_prop_yes = age_merged[age_merged.y=='yes']
age_prop_yes.head()
```

```python
age_prop_yes = age_merged[age_merged.y=='yes']
age_prop_no = age_merged[age_merged.y=='no']
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.plot(age_prop_yes['age'], age_prop_yes['age_prop'])
plt.xlabel('Yes')
plt.ylabel('Age')
plt.subplot(1, 2, 2)
plt.plot(age_prop_no['age'], age_prop_no['age_prop'])
plt.xlabel('No')
plt.ylabel('Age');
```
